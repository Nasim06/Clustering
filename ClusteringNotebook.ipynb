```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

Iris_data contains a 2D array. 150 samples. 4 features: Sepal length, Sepal width, Petal width and Petal length. 
Iris_labels contains a 1D of 150 samples. These are the clusters for the Iris data split into 3 groups: 0, 1 and 2.

```python
iData = np.load('./Iris_data.npy')
iLabels = np.load('./Iris_labels.npy')

x_feature = 0
y_feature = 1

plt.figure()
plt.scatter(iData[:, x_feature], iData[:, y_feature], c=iLabels)
plt.title('Iris flower dataset')
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.grid()
plt.show()
```

Task 2.2

The following is an attempt to cluster these data points using K mean clustering. I use the KMeans class to predict the cluster memberships and matplotlib to plot this data with centroids so I can visually compare it to the correct clusters.

```python
kmeans = KMeans(n_clusters=3, init = 'k-means++')
KMmodel = kmeans.fit(iData)
prediction = kmeans.predict(iData)
```

```python
plt.figure
plt.scatter(iData[:, 0], iData[:, 1], c=prediction, s=20, cmap='viridis');
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=150, marker='*', c='red', edgecolor='black')
plt.title('Iris flower dataset')
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.grid()
plt.show()
```

Task 2.3

```python
GM = GaussianMixture(n_components = 3)
GM.fit(iData)
label = GM.predict(iData)
means = GM.means_
```

```python
plt.figure()
plt.scatter(iData[:, 0], iData[:, 1], c=label, s=20, cmap='viridis');
plt.scatter(means[:, 0], means[:, 1],s=150, marker='*', c='red', edgecolor='black')
plt.title('Iris flower dataset')
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.grid()
plt.show()
```

```python
pp = GM.predict_proba(iData)
plt.figure()
plt.scatter(iData[:, 0], iData[:, 1], c=pp, s=20, cmap='viridis');
plt.title('Iris flower dataset')
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.grid()
plt.show()
```
